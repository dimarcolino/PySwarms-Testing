# Função Rosenbrock

```python
# create a parameterized version of the classic Rosenbrock unconstrained optimzation function
def rosenbrock_with_args(x, a, b, c=0):
    f = (a - x[:, 0]) ** 2 + b * (x[:, 1] - x[:, 0] ** 2) ** 2 + c
    return f
```

## Usando GlobalBest

```python
# instatiate the optimizer
x_max = 10 * np.ones(2)
x_min = -1 * x_max
bounds = (x_min, x_max)

options = {'c1': 0.5, 'c2': 0.3, 'w': 0.9}
optimizer = ps.single.GlobalBestPSO(n_particles=70, dimensions=2, options=options, bounds=bounds)

# now run the optimization, pass a=1 and b=100 as a tuple assigned to args

cost, pos = optimizer.optimize(rosenbrock_with_args, 200, a=1, b=100, c=0)
```

```python
plot_cost_history(cost_history=optimizer.cost_history)
plt.show()
```

```python
m = Mesher(func=fx.rosenbrock, limits=[(-1,2), (-1,2)])
```

```python
%%capture
# Make animation
fig, ax = plt.subplots(figsize=(10, 5))
animation_rosenbrock = plot_contour(pos_history=optimizer.pos_history,
                         mesher=m,
                         mark=(1,1),
                         canvas=(fig, ax))
ax.set_xlim(-1, 2)
ax.set_ylim(-1, 2)
```

```python
animation_rosenbrock.save('plot_rosenbrock.gif', writer='imagemagick', fps=10)
Image(url='plot_rosenbrock.gif')
```
